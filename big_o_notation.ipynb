```python
import pandas as pd
import matplotlib.pyplot as plt
print(plt)

from random import randint
import math
from math import log
import numpy as np
```

```python
a = pd.Series([2, 3, 4, 5, 6])

def is_three_or_five(x):
    return x == 3 or x == 5

print(is_three_or_five(10)) # single number function input
a.apply(is_three_or_five) # series.apply(function_name)
```

```python
x = pd.Series(range(1, 14))
y = 0*x + 1
```

```python
y1 = np.log(x)
```

```python
y2 = x
```

```python
y3 = x * np.log(x)
```

```python
y4 = x*x
```

```python
y5 = 2**x
```

```python
y6 = x.apply(np.math.factorial)
```

```python
y7 = x**x
```

```python
# 5. Title your chart 
#e. y= (0*n) +1 and label the curve "O(1)"
plt.plot(x, y, label="0(n) + 1")
plt.plot(x, y1, label="log(n)")
plt.plot(x, y2, label="y = n")
plt.plot(x, y3, label="n * log(n)")
plt.plot(x, y4, label="x**2")
plt.plot(x, y5, label="2**x")
plt.plot(x, y6, label="factorial(x)")
#a Title your chart "Big O Notation"
plt.title("Big O Notation")
#b. plt.xlabel('$x$')
plt.xlabel('$x$')
#c Label your y axis "Operations"
plt.ylabel('Operations')         
#d Label your curves or make a legend for the curves  --   
plt.xlim(1,13)
plt.ylim(0,20)
plt.legend(loc='upper right')
plt.show()
```
